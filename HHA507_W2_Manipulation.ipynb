### Handling Missing Values in Dataset

```python
# Check for Missing Values in Columns
missing_values = df.isnull().sum()
print(missing_values)
```

### Removing White Space and Special Characters

```python
import re
import numpy as np
```

```python
# Function to remove white space and special characters from a value
def clean_value(value):
  cleaned_value = re.sub(r'\s+|[^a-zA-z0-9]', '', value)
  return cleaned_value

# Apply the clean_value function to all the columns
for column in df.columns:
  df[column] = df[column].apply(clean_value)

df
```

### Identifying Duplicate Rows and Columns

```python
# Identify and remove duplicate rows
df.drop_duplicates(inplace=True)
df
```

### Cleaning Column Names

```python
df.columns = df.columns.str.strip()
df
```

```python
# Convert 'Patient Age' and 'Average heart Rate' to numeric values
df['Patient Age'] = pd.to_numeric(df['Patient Age'], errors='coerce')
df['Average Heart Rate'] = pd.to_numeric(df['Average Heart Rate'], errors='coerce')

bins = [0, 30, 50, np.inf]
labels = ['Young', 'Middle-Aged', 'Senior']

# Creating New Column Called 'Age Group'
df['Age Group'] = pd.cut(df['Patient Age'], bins=bins, labels=labels, right=False)
```

```python
df['Age Group']
```

```python
# Creating a Pivot Table
avg_patientHeartRate_Pivot = df.pivot_table(values='Average Heart Rate', index='Gender', columns='Age Group', aggfunc='mean')

# Display Pivot Table
avg_patientHeartRate_Pivot
```

```python
df['Disease Diagnosed']
```

```python
# Convert Average BP to numeric
df['Average BP'] = df['Average BP'].astype(str).astype(float)

#Group by Average BP based on disease diagnosed
grouped_data = df.groupby('Disease Diagnosed').agg({'Average BP': 'mean'})
print(grouped_data)
```
