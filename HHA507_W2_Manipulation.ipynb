### Identifying and Deleting Duplicate Rows and Columns

```python
# Identify Duplicate Reecords
print(df.duplicated())
```

### Cleaning Column Names

```python
df.columns = df.columns.str.strip()
df
```

### Removing White Space and Special Characters

```python
# Function to remove white space and special characters from a value
def clean_column_names(data):
  def clean_name(value):
    cleaned_name = re.sub(r'[^a-zA-Z0-9$()]', '_', value)
    return cleaned_name.lower()

# Applying function for all columns
  data.columns = [clean_name(col) for col in data.columns]
  return data

df = clean_column_names(df)
```

### Creating a New Column

```python
df.dtypes
```

```python
# Convert 'payment_due_(\$)' and 'prescription_cost_(\$)' to numeric values
df['payment_due_($)'] = pd.to_numeric(df['payment_due_($)'], errors='coerce')
df['prescription_cost_($)'] = pd.to_numeric(df['prescription_cost_($)'], errors='coerce')

#Creating New Column called 'NonPrescription_cost'
NonPrescription_cost = (df['payment_due_($)']-df['prescription_cost_($)'])
```

### Aggregating Data Using groupby and Compute Summary Statistics

```python
# Generate descriptive statistics for numerical columns
print(df.describe())
```

```python
Payment = df[['has_insurance', 'payment_method', 'payment_due_($)', 'prescription_cost_($)']]
Payment.head(15)
```

```python
# Convert 'patient_age' and 'average_heart_rate' to numeric values
df['patient_age'] = pd.to_numeric(df['patient_age'], errors='coerce')
df['average_heart_rate'] = pd.to_numeric(df['average_heart_rate'], errors='coerce')

bins = [0, 30, 50, np.inf]
labels = ['Young', 'Middle-Aged', 'Senior']

# Creating New Column Called 'age_group'
df['age_group'] = pd.cut(df['patient_age'], bins=bins, labels=labels, right=False)
```

```python
df['age_group']
```

```python
# Creating a Pivot Table
avg_patientHeartRate_Pivot = df.pivot_table(values='average_heart_rate', index='gender', columns='age_group', aggfunc='mean')

# Display Pivot Table
avg_patientHeartRate_Pivot
```

```python
df['disease_diagnosed']
```

```python
# Convert 'average_bp' to numeric
df['average_bp'] = pd.to_numeric(df['average_bp'], errors='coerce')

#Group by 'averagebp' based on 'disease_diagnosed'
grouped_data = df.groupby('disease_diagnosed').agg({'average_bp': 'mean'})
print(grouped_data)
```

### Advanced Exercises

```python
import pandas as pd
import time
```

```python
# Define the dataset path
dataset_path = '/content/diabetes.csv'
```

```python
df3 = pd.read_csv('/content/diabetes.csv')
column_data_type_df3 = df3.dtypes
print(column_data_type_df3)
```

#### Check Load Time Without Specific Data

```python
# Using Pandas
start_time = time.time()
pandas_df3 = pd.read_csv(dataset_path)
pandas_time = time.time() - start_time
```

```python
# Using Polar
start_time = time.time()
polar_df3 = pd.read_csv(dataset_path)
polar_time = time.time() - start_time
```

```python
# Using Modin
start_time = time.time()
modin_df3 = pd.read_csv(dataset_path)
modin_time = time.time() - start_time
```

```python
# Results and Execution Times
print(f"Pandas Execution Time: {pandas_time:.4f} seconds")
print(f"Polar Execution Time: {polar_time:.4f} seconds")
print(f"Modin Execution Time: {modin_time:.4f} seconds")
```

#### Check Load Time with Specific Data

```python
# Using Pandas
start_time = time.time()
pandas_df3 = pd.read_csv(dataset_path)
pandas_filtered = pandas_df3[pandas_df3['Outcome'] == 1]
pandas_average_glucose = pandas_filtered['Glucose'].mean()
pandas_time = time.time() - start_time
```

```python
# Using Polar
start_time = time.time()
polar_df3 = pd.read_csv(dataset_path)
polar_filtered = polar_df3[polar_df3['Outcome' ]== 1]
polar_average_glucose = polar_filtered['Glucose'].mean()
polar_time = time.time() - start_time
```

```python
# Using Modin
start_time = time.time()
modin_df3 = pd.read_csv(dataset_path)
modin_filtered = modin_df3[modin_df3['Outcome'] == 1]
modin_average_glucose = modin_filtered['Glucose'].mean()
modin_time = time.time() - start_time
```

```python
# Results and Execution Times
print(f"Pandas Average Glucose with Positive Outcome: {pandas_average_glucose:.2f} mmol/l")
print(f"Polar Average Glucose with Positive Outcome: {pandas_average_glucose:.2f} mmol/l")
print(f"Modin Average Glucose with Positive Outcome: {modin_average_glucose:.2f} mmol/l")
print(f"Pandas Execution Time: {pandas_time:.4f} seconds")
print(f"Polar Execution Time: {polar_time:.4f} seconds")
print(f"Modin Execution Time: {modin_time:.4f} seconds")
```

#### Summary of Load Times

* The load time of the Modin is always the fastest, then the load time of polar, then pandas.
