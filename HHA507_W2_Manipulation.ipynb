### Removing White Space and Special Characters

```python
# Function to remove white space and special characters from a value
def clean_column_names(data):
  def clean_name(value):
    cleaned_name = re.sub(r'[^a-zA-Z0-9$()]', '_', value)
    return cleaned_name.lower()

# Applying function for all columns
  data.columns = [clean_name(col) for col in data.columns]
  return data

df = clean_column_names(df)
```

### Creating a New Column

```python
df.dtypes
```

```python
# Convert 'Payment Due ($)' and 'Prescription Cost ($)' to numeric values
df['payment_due_($)'] = pd.to_numeric(df['payment_due_($)'], errors='coerce')
df['prescription_cost_($)'] = pd.to_numeric(df['prescription_cost_($)'], errors='coerce')

#Creating New Column called 'NonPrescription_cost'
NonPrescription_cost = (df['payment_due_($)']-df['prescription_cost_($)'])
```

```python
NonPrescription_cost
```

### Aggregating Data Using groupby and Compute Summary Statistics

```python
# Generate descriptive statistics for numerical columns
print(df.describe())
```

```python
# Convert 'Patient Age' and 'Average heart Rate' to numeric values
df['patient_age'] = pd.to_numeric(df['patient_age'], errors='coerce')
df['average_heart_rate'] = pd.to_numeric(df['average_heart_rate'], errors='coerce')

bins = [0, 30, 50, np.inf]
labels = ['Young', 'Middle-Aged', 'Senior']

# Creating New Column Called 'Age Group'
df['age_group'] = pd.cut(df['patient_age'], bins=bins, labels=labels, right=False)
```

```python
df['age_group']
```

```python
# Creating a Pivot Table
avg_patientHeartRate_Pivot = df.pivot_table(values='average_heart_rate', index='gender', columns='age_group', aggfunc='mean')

# Display Pivot Table
avg_patientHeartRate_Pivot
```

```python
df['disease_diagnosed']
```

```python
# Convert Average BP to numeric
df['average_bp'] = pd.to_numeric(df['average_bp'], errors='coerce')

#Group by Average BP based on disease diagnosed
grouped_data = df.groupby('disease_diagnosed').agg({'average_bp': 'mean'})
print(grouped_data)
```
